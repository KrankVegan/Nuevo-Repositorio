```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python

def load_image(path_image):
    img = cv2.imread(path_image)
    return img[...,::-1] #BGR to RGB

img = load_image('20221013_093355.jpg')

print(f'Alto : {img.shape[0]}')
print(f'Ancho : {img.shape[1]}')
print(f'Canales : {img.shape[2]}')

fig, axs = plt.subplots(1,4,figsize=(20,5))

axs[0].imshow(img)
axs[0].set_title('Imagen')


fig_img = axs[1].imshow(img[...,0])
fig.colorbar(fig_img, ax=axs[1],shrink=0.5)
axs[1].set_title('Canal Rojo')


fig_img = axs[2].imshow(img[...,1])
fig.colorbar(fig_img, ax=axs[2],shrink=0.5)
axs[2].set_title('Canal Verde')


fig_img = axs[3].imshow(img[...,2])
fig.colorbar(fig_img, ax=axs[3],shrink=0.5)
axs[3].set_title('Canal Azul');
```

```python
hsv_cube = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)#se pasa a HSV para manejar mas facil
```

```python
#Rango de Colores
#all hsv values
light_orange = (1, 190, 200)
dark_orange = (18, 255, 255)

light_blue = (94, 80, 2)
dark_blue = (126, 255, 255)

light_green =	(70,0,0)
dark_green =	(86,255,255)

light_yellow = (22, 93, 0)
dark_yellow = (45, 255, 255)

light_red = (0,50,50)
dark_red = (2,255,255)

```

```python
fig, axs = plt.subplots(1,6,figsize=(27,5))

axs[0].imshow(img)
axs[0].set_title('Imagen')

mask = cv2.inRange(hsv_cube, light_orange, dark_orange)
result_o = cv2.bitwise_and(img,img, mask=mask)

fig_img = axs[1].imshow(result_o)
axs[1].set_title('Mascara Naranja')

mask = cv2.inRange(hsv_cube, light_green, dark_green)
result_g = cv2.bitwise_and(img,img, mask=mask)


fig_img = axs[2].imshow(result_g)
axs[2].set_title('Mascara Verde')

mask = cv2.inRange(hsv_cube, light_yellow, dark_yellow)
result_y = cv2.bitwise_and(img,img, mask=mask)


fig_img = axs[3].imshow(result_y)
axs[3].set_title('Mascara Amarilla')

mask = cv2.inRange(hsv_cube, light_blue, dark_blue)
result_b = cv2.bitwise_and(img,img, mask=mask)


fig_img = axs[4].imshow(result_b)
axs[4].set_title('Mascara Azul')

mask = cv2.inRange(hsv_cube, light_red, dark_red)
result_r = cv2.bitwise_and(img,img, mask=mask)


fig_img = axs[5].imshow(result_r)
axs[5].set_title('Mascara Roja')
```
